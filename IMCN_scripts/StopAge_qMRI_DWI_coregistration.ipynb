# Coregistration of qMRI to B0
Bringing individual myelin and iron maps into the subject's DWI space with rigid registration

```python
import nighres
import nibabel
import numpy
import glob
```

```python
dwi_dir = '/home/public/StoppingAge-DWI-Ahead/Data/Dwi/'
qmr_dir = '/home/public/StoppingAge-DWI-Ahead/Data/qMRI-estimates/'
```

```python
qmr_myelin_files = glob.glob(qmr_dir+'sub-*/*-myelin.nii.gz')
```

```python
for qmr_file in qmr_myelin_files:
    # get subject id
    sub_id = qmr_file.replace(qmr_dir,'')[0:7]
    print('subject: '+sub_id)
    # find corresponding B0 map
    b0_file = glob.glob(dwi_dir+sub_id+'/preprocessed/mean_b0.nii.gz')
    if (len(b0_file)==0): print('mean B0 not found!')
    else:
        b0_file = b0_file[0]
        out_dir = dwi_dir+sub_id+'/qmri_est/'
        
        coreg = nighres.registration.embedded_antsreg(source_image=qmr_file,target_image=b0_file,
                                                      run_rigid=True,run_affine=False,run_syn=False,
                                                      save_data=True,output_dir=out_dir)
        
        qmr_iron = glob.glob(qmr_dir+sub_id+'/*-iron.nii.gz')
        if (len(qmr_iron)==0): print("missing iron file!")
        else:
            qmr_iron = qmr_iron[0]
            trans = nighres.registration.apply_coordinate_mappings(image=qmr_iron,mapping1=coreg['mapping'],
                                                                  save_data=True,output_dir=out_dir)
```
